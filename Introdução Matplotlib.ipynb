# Two options for subplots


```python
#Subplot option 1(Recommended)
fig, ((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(energicPrices.keys(),energicPrices.values());
ax4.hist(np.random.randn(1000));
```

```python
#subplot option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot to each differente index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(energicPrices.keys(),energicPrices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Wich one should you use? (pyplot vs matplotlib OO method?)
- When plotting somethin quiclik, okay to use pyplot but matplot is better
- when using some complex use OO method.



```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO method mixed with pyplot

fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol", c="target", ax=ax);
#o ax colocou o age ali que antes n tinha.
#ax.set_xlim([45,100]); Isso faz a distancia do age ficar entre 45 e 100
#sem rodar isso ele fica 50 para 75 como ta ai em baixo
```

```python
#OO method from scratch
fig,ax= plt.subplots(figsize=(10,6))
#plot the data
scatter= ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
#O C ele ta trocando a cor dependendo da informacao que a coluna tiver.
#Customize plot:
ax.set(title="Heart diseases",xlabel="Age",ylabel="Cholesterol")
#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
#Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
#Subplot of chol ,age, thalach
fig, (ax0, ax1)= plt.subplots(nrows=2, ncols=1, figsize=(10,10),sharex=True )
#Sharex compartilha o cdigo entre os dois AX
#Add gdata to ax0
scatter= ax0.scatter(x=over_50["age"],y=over_50["chol"], c=over_50["target"])
#Customize ax0
ax0.set(title="Heart disease and cholesterol", ylabel="Cholesterol")
#Add a legend
ax0.legend(*scatter.legend_elements(),title="Target")
#Add a meanline
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

#Add data to ax1
scatter= ax1.scatter(x=over_50["age"],y=over_50["thalach"], c=over_50["target"]);
ax1.set(title="Max heart rate", xlabel="Age",ylabel="Max heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");

#Add a title of a figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

# matplot lib colors has all the colors customization

```python
#Customizing the y and x axis limitations

#Subplot of chol ,age, thalach
fig, (ax0, ax1)= plt.subplots(nrows=2, ncols=1, figsize=(10,10),sharex=True )
#Sharex compartilha o cdigo entre os dois AX
#Add gdata to ax0
scatter= ax0.scatter(x=over_50["age"],y=over_50["chol"], c=over_50["target"], cmap="winter")
#Customize ax0
ax0.set(title="Heart disease and cholesterol", ylabel="Cholesterol")
#Change the x axis limits
ax0.set_xlim([50,80])
#Change the y axis limits



#Add a legend
ax0.legend(*scatter.legend_elements(),title="Target")
#Add a meanline
ax0.axhline(y=over_50["chol"].mean(),color='r', 
            linestyle='--', 
            label="Average");

#Add data to ax1
scatter= ax1.scatter(x=over_50["age"],y=over_50["thalach"], c=over_50["target"],cmap="winter");
ax1.set(title="Max heart rate", xlabel="Age",ylabel="Max heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(),color='r', 
            linestyle='--', 
            label="Average");
ax1.set_ylim([60,200])

#nao precisa limitar o  x axis pq o que ta la em cima j serve dos dois
#pq eles compartilham o mesmo X axis. no caso de 50 a 80 que est.


#Add a title of a figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
